```python
import pandas as pd 
import numpy as np 
```

```python
match = pd.read_csv("matches.csv")
delivery = pd.read_csv("deliveries.csv")
```

```python
match.head(2)
```

```python
delivery.head(2)
```

```python
df = delivery.groupby("batsman")["batsman_runs"].sum().reset_index(name = "Total run")
```

```python
df
```

```python
df2 = delivery.groupby("batsman")["match_id"].count().reset_index(name="Total Ball")
```

```python
df2
```

```python
df3 = df.merge(df2,on="batsman",how="inner")
```

```python
df3["ratio_ball_runs"] = df3["Total run"]/df3["Total Ball"]
```

```python
df3.sort_values(by="ratio_ball_runs",ascending=False).head(10)
```

```python
df3[df3['Total Ball']>500].sort_values(by="ratio_ball_runs",ascending=False).head(10)
```
