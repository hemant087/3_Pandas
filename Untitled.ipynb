```python
# 60. How to create a new column that contains the row number of nearest column by euclidean distance?
# Create a new column such that, each row contains the row number of nearest row-record by euclidean distance.

# DifficultyÂ Level:Â L3
```

```python
import pandas as pd 
import numpy as np 
np.random.seed(100)
df = pd.DataFrame(np.random.randint(1,100, 40).reshape(10, -1), columns=list('pqrs'), index=list('abcdefghij'))
df
```

```python
# init outputs
nearest_rows = []
nearest_distance = []

# iterate rows.
for i, row in df.iterrows():
    curr = row
    rest = df.drop(i)
    e_dists = {}  # init dict to store euclidean dists for current row.
    # iterate rest of rows for current row
    for j, contestant in rest.iterrows():
        # compute euclidean dist and update e_dists
        e_dists.update({j: round(np.linalg.norm(curr.values - contestant.values))})
    # update nearest row to current row and the distance value
    nearest_rows.append(max(e_dists, key=e_dists.get))
    nearest_distance.append(max(e_dists.values()))

df['nearest_row'] = nearest_rows
df['dist'] = nearest_distance
```

```python
from scipy.spatial.distance import cdist

# Compute the distance matrix
dist_matrix = cdist(df.values, df.values, metric='euclidean')

# Replace diagonal with infinity to avoid matching the same row
np.fill_diagonal(dist_matrix, np.inf)

# Find the index (row number) of the nearest neighbor
nearest_indices = np.argmin(dist_matrix, axis=1)

# Create a new column with the row label of the nearest neighbor
df['nearest'] = df.index[nearest_indices]

print(df)
```

```python
# Step 1: Convert DataFrame to NumPy array
data = df.values

# Step 2: Compute pairwise Euclidean distances using broadcasting
diff = data[:, np.newaxis, :] - data[np.newaxis, :, :]  # shape: (10, 10, 4)
dists = np.sqrt(np.sum(diff**2, axis=2))  # shape: (10, 10)

# Step 3: Ignore distance to self by setting diagonal to infinity
np.fill_diagonal(dists, np.inf)

# Step 4: Get the index of the nearest row for each row
nearest_indices = np.argmin(dists, axis=1)

# Step 5: Add the row labels of the nearest rows as a new column
df['nearest'] = df.index[nearest_indices]

print(df)
```

```python
def pointTable(season):
    ipl = df[df["Season"] == season]
    ipl = ipl.iloc[:,[5,6,11,12]]
    NoR = ipl[ipl.WonBy == "NoResults"]
    T1 = ipl['Team1'].value_counts()
    T2 = ipl['Team2'].value_counts()
    Won = ipl["WinningTeam"].value_counts()
    
    point = pd.DataFrame({"T1_count":T1,"T2_count":T2,"Won_count":Won})
    point["total_count"] = point['T1_count']+point['T2_count']
    point["Points"] = point["Won_count"]*2
    point["NoResult"] = np.nan
    for i,val in NoR.iterrows():
        point.loc[val["Team1"],"Points"]+=1
        point.loc[val["Team1"],"NoResult"]=1
        point.loc[val["Team2"],"Points"]+=1
        point.loc[val["Team2"],"NoResult"]=1
    return point
```
