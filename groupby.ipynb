```python
# find the top 10 batsman in terms of runs
ipl.groupby("batsman")["batsman_runs"].sum().sort_values(ascending = False).head(10)
```

```python
# find the batsman with max no of sixes
ipl[ipl["batsman_runs"]==6].groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head(5)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
ipl[(ipl["over"]>=15)&((ipl["batsman_runs"]==6)|(ipl["batsman_runs"]==4))].groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head(5)
```

```python
# find V Kohli's record against all teams
ipl[ipl["batsman"]=="V Kohli"].groupby(["bowling_team"])["batsman_runs"].sum()
```

```python
# Create a function that can return the highest score of any batsman
def high_score(batsman):
    ipl[ipl["batsman"]== batsman]
    hig = ipl[ipl["batsman"]== batsman].groupby("match_id")["batsman_runs"].sum()
    return hig.sort_values().head(1).index[0]
```
