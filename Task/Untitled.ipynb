###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

```python
# count = ipl.batter.value_counts()
df = pd.DataFrame({"total_run":ipl[ipl['innings']==2].groupby("batter")['batsman_run'].sum(),"Total_ball":ipl.batter.value_counts()})
```

```python
df.head()
```

```python
df["strike_rate"] = (df["total_run"]/df["Total_ball"])*100
```

```python
df[df["Total_ball"]>100].sort_values("strike_rate",ascending=False)
```
